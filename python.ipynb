```python
base_url = 'https://github.com/topics/python?page='
```

```python
url = []
for num in range(1,8):

    url.append(base_url+str(num))
```

```python
url
```

```python
all_article_lst = []
for url_data in url:
    html_data = requests.get(url_data).text
   # print(url_data)
    converted_data = BeautifulSoup(html_data,'lxml')
    all_article = converted_data.find_all('article',class_='border rounded color-shadow-small color-bg-subtle my-4')
   # print(all_article)
    for item in all_article:
        Repository_name = item.find('a',class_='text-bold wb-break-word').text
        Repository_name = Repository_name.strip()
        User_name = item.find('a',class_='').text
        User_name = User_name.strip()
        Stars = item.find('span',class_='Counter js-social-count').text
        URL = item.find('a',class_='text-bold wb-break-word')['href']
        URL = URL.strip()
        
        
        dict2 = {
            'Repository Name': Repository_name,
            'User Name': User_name,
            'Stars': Stars,
            'URL': base_Url +URL
        }
        
        print(dict2)
        
        
        all_article_lst.append(dict2)
```

```python
import pandas as pd
df = pd.DataFrame(all_article_lst)
```
